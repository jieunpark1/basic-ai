## 다중 분류를 위해 알아야 하는 것.
- 2개 이상, 3개 이상의 범주가 있을 때 어떻게 분류하는지? (다중분류)
- 아이리스 꽃 분류 데이터가 가장 많이 사용됨
## *one-hot encoding*알고리즘
- 3개 종류 setosa, versicolor, virginica 중 해당되는 것을 1로, 아닌것을 0으로 표기.
## **softmax**
- 각 카테고리(클래스)를 0과 1사이의 확률로 표현하여 모든 카테고리 확률의 합이 1이 되도록 만드는 함수
- 가장 높은 확률의 카테고리로 모델이 예측.

## 원-핫 인코딩
원-핫 인코딩에 대해 배워 봅시다.

<img src="https://github.com/taehojo/fastcampus_ai/blob/master/data/img/03-02.png?raw=1" alt="dl-structure" width="550"/>

```python
# 속성을 X, 클래스를 y로 저장합니다.
X = df.iloc[:,0:4]
y = df.iloc[:,4]

# X와 y의 첫 5줄을 출력해 보겠습니다.
print(X[0:5])
print(y[0:5])
```

```python
# 원-핫 인코딩 처리를 합니다.
y = pd.get_dummies(y)

# 원-핫 인코딩 결과를 확인합니다.
print(y[0:5])
```

## one hot encoding 데이터 만드는 방법-> pd.get_dummies(y)

## 소프트맥스
소프트 맥스를 적용해 봅시다.  

<img src="https://github.com/taehojo/fastcampus_ai/blob/master/data/img/03-03.png?raw=1" width="550"/>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 모델 설정
model = Sequential()
model.add(Dense(12,  input_dim=4, activation='relu'))
model.add(Dense(8,  activation='relu'))
model.add(Dense(3, activation='softmax')) #마지막 output layer에 3개 클래스에 대해 softmax activation function 적용 (instead of binary-cross-entropy)
model.summary()

# 모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) #loss function => categorical crossentropy instead of binary cross entropy.

# 모델 실행
history=model.fit(X, y, epochs=30, batch_size=5)
```

## [전체] 아이리스 품족 예측 실행

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 깃허브에 준비된 데이터를 가져옵니다. 앞에서 이미 가져왔으므로 주석 처리합니다. 5번 예제만 별도 실행 시 주석을 해제한 후 실행해주세요.
# !git clone https://github.com/taehojo/data.git

# 아이리스 데이터를 불러옵니다.
df = pd.read_csv('./data/iris3.csv')

# 속성을 X, 클래스를 y로 저장합니다.
X = df.iloc[:,0:4]
y = df.iloc[:,4]

# 원-핫 인코딩 처리를 합니다.
y = pd.get_dummies(y)

# 모델 설정
model = Sequential()
model.add(Dense(12,  input_dim=4, activation='relu'))
model.add(Dense(8,  activation='relu'))
model.add(Dense(3, activation='softmax'))
model.summary()

# 모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 실행
history=model.fit(X, y, epochs=30, batch_size=5)
```

과적합 -> 학습셋/테스트셋 따로 준비
- 학습셋은 에폭 지날수록 에러 줄어듦
- 테스트셋은 에폭 지날수록 에러 높아진다 -> 이게 과적합

-교차검증 : 데이터셋 별로 없을때 (cross validation) 1/5로 데이터를 나누고, 랜덤 1세트는 테스트셋으로 설정하고 나머지 4세트는 훈련 세트로하며 5번 반복하며 훈련하는 방법
