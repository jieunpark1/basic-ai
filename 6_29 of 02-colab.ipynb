이번에는 피마 인디언의 당뇨병 데이터로 실행해 봅시다.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# pandas 라이브러리를 불러옵니다.
import pandas as pd

# 깃허브에 준비된 데이터를 가져옵니다. 이미 앞에서 가져왔으므로 주석 처리합니다. 5번 예제만 별도 실행 시 주석 해제 후 실습하세요.
# !git clone https://github.com/taehojo/data.git

# 피마 인디언 당뇨병 데이터셋을 불러옵니다.
df = pd.read_csv('./data/pima-indians-diabetes3.csv')
```

```python
print(df.head)
```

```python
# 세부 정보를 X로 지정합니다.
X = df.iloc[:,0:8]
# 당뇨병 여부를 y로 지정합니다.
y = df.iloc[:,8]
```

위 구조를 그림으로 표현하면 다음과 같습니다.

<img src="https://github.com/taehojo/fastcampus_ai/blob/master/data/img/02-02.png?raw=1" alt="dl-structure" width="700"/>

```python
# 모델을 설정합니다.
model = Sequential()  #첫번째 Input layer는 설정 안함.
model.add(Dense(12, input_dim=8, activation='relu', name='1st_Layer')) #(은닉층 dim, input_dim, activation function) #첫번쨰 은닉층
model.add(Dense(8, activation='relu', name='2nd_Layer')) # 8->8 #은닉층2번쨰
model.add(Dense(1, activation='sigmoid',name='3rd_Layer')) # 8->1 #sigmoid를 써야 0에 가까운지 1에 가까운지 분류문제에서 많이 씀.
model.summary()
```

 - (input -> hid1) weight: 8*12=96 / bias(hid하나씩) 12개 = 96+12 = 108개 계산해야할 파라미터 수

 - (hid1 -> hid2) weight 12*8 + bias 8 = 96 + 8 = 104개 파라미터

 - (hid2 -> output) weight 8*1 + bias 1 = 8 + 1 = 9개 파라미터

 -> 총 108 + 104 + 9 = 221개 파라미터 학습 필요



```python
# 모델을 컴파일합니다.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) # loss 함수 : 분류 -> binary cross-entropy, 경사하강법,

# 모델을 실행합니다.
history=model.fit(X, y, epochs=100, batch_size=5) #전체 샘플 460개 한번에 아니고 샘플 5개씩만. 한 epoch에서 훈련. 큰 데이터는 한번에 넣지 않음 cost 커져서.
```
